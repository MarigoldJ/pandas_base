# Get url and Request data
[공공데이터포털 바로가기](https://data.go.kr/tcs/dss/selectApiDataDetailView.do?publicDataPk=15043376)

```python
# api key
key = input('Write API key > ')
url_base = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson'

query = {
    quote_plus('pageNo'): 1,
    quote_plus('numOfRows'): 10,
    quote_plus('startCreateDt'): '20200320',
    quote_plus('endCreateDt'): '20210526'
}

option = urlencode(query)
# key = '1'
# url for request
url = f'{url_base}?serviceKey={key}&{option}'

# response
print('Data load... ', end='')

response = urlopen(url)
results = response.read().decode('utf-8')
results_json = xmltodict.parse(results)

data = json.loads(json.dumps(results_json))

if 'ERROR' in data['response']['header']['resultMsg']:
    print('Fail.')
else:
    print('Success!')
```

```python
logs = data['response']['body']['items']['item']
df = pd.DataFrame()

for log in logs:
    s_tmp = pd.Series(log)
    df_tmp = pd.DataFrame(s_tmp)
    df = pd.concat([df, df_tmp.T])

# column 명칭 알기 쉽게 바꾸기
col_names = {
    'accDefRate': '누적 환진률',
    'accExamCnt': '누적 검사 수',
    'accExamCompCnt': '누적 검사 완료 수',
    'careCnt': '치료중 환자 수',
    'clearCnt': '격리해제 수',
    'createDt': '등록일시분초',
    'deathCnt': '사망자 수',
    'decideCnt': '확진자 수',
    'examCnt': '검사진행 수',
    'resutlNegCnt':'결과 음성 수',
    'seq': '게시글번호(감염현황 고유값)',
    'stateDt': '기준일',
    'stateTime': '기준시간',
    'updateDt': '수정일시분초',
}
df.rename(columns = col_names, inplace=True)
df
```
