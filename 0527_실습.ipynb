```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]
df.head()
```

```python
def missing_value(x):
    return x.isna()

def missing_count(x):
    return missing_value(x).sum()

def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result_df = df.pipe(total_number_missing)
result_df
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
mask1 = (df.age >= 10) & (df.age < 20)
mask1.head()
```

```python
df_teenage = df.loc[mask1, :]
df_teenage.head()
```

```python
# 10살 미만, 남성인 승객 선택하기
mask_custom = (df.age < 10) & (df.sex == 'male')
df_custom = df.loc[mask_custom, :]
df_custom.head()
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택
mask2 = (df.age < 10) | (df.age >= 60)
col2 = ['age', 'sex','alone']
df_2 = df.loc[mask2, col2]
df_2
```

```python
# 함께 탑승한 형제 또는 배우자의 수가 3, 4, 5인 승객만 따로 추출
a = (df['sibsp'] == 3)
b = (df['sibsp'] == 4)
c = (df['sibsp'] == 5)
df_boolean = df[a|b|c]
df_boolean.head()
```

```python
# .isin()
isin_filter = df['sibsp'].isin([3, 4, 5])
df_boolean = df[isin_filter]
df_boolean.head()
```

```python
import pandas as pd
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])

print(df1, '\n\n')
print(df2)
```

```python
result1 = pd.concat([df1, df2], ignore_index=True)
result1
```

```python
result1 = pd.concat([df1, df2], axis=1, ignore_index=True)
result1
```

```python
result1 = pd.concat([df1, df2], axis=1, join='inner') # 교집합
result1
```

```python
result1 = pd.concat([df1, df2], axis=1, join='outer') # 합집합
result1
```
